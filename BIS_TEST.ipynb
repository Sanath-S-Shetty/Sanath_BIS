```python
import random


def calculate_fitness(gen, target):

    matches = sum([1 for i in range(len(target)) if gen[i] == target[i]])
    return (matches / len(target)) * 100

def create_population(target_length, population_size, genes):
    population = []
    for _ in range(population_size):
        chromosome = [random.choice(genes) for _ in range(target_length)]
        population.append("".join(chromosome))
    return population


def selection_and_evolution(population, target, mutation_rate, genes):

    population_with_fitness = sorted([(chromo, calculate_fitness(chromo, target)) for chromo in population], key=lambda x: x[1], reverse=True)


    parents = [item[0] for item in population_with_fitness[:len(population_with_fitness)//2]]


    next_generation = []
    for _ in range(len(population)):
        parent1 = random.choice(parents)
        parent2 = random.choice(parents)
        crossover_point = random.randint(1, len(target) - 1)

        child = parent1[:crossover_point] + parent2[crossover_point:]


        child_list = list(child)
        for i in range(len(child_list)):
            if random.random() < mutation_rate:
                child_list[i] = random.choice(genes)
        child = "".join(child_list)
        next_generation.append(child)

    return next_generation

print("enter the key")
target = input()
genes = "abcdefghijklmnopqrstuvwxyz"
population_size = 100
mutation_rate = 0.1

population = create_population(len(target), population_size, genes)

for generation in range(1000):
    population = selection_and_evolution(population, target, mutation_rate, genes)
    best_solution = max([(chromo, calculate_fitness(chromo, target)) for chromo in population], key=lambda x: x[1])
    print(f"Generation {generation}: Best solution '{best_solution[0]}' with fitness {best_solution[1]:.2f}%")
    if best_solution[1] == 100:
        print("Target found!")
        break
```
